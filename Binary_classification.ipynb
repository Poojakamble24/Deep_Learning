**### step 4: Perform  preprocessing on X**

```python
X.dtypes
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
from re import S
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),('scalar',StandardScaler())]).set_output(transform='pandas')
```

```python
num_pipe
```

```python
pre=num_pipe.fit_transform(X)
```

```python
pre
```

```python
### step 5 Train Test spli
```

**### step 5 Train Test split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(pre,Y,test_size=0.2,random_state=23)
```

**### building a neural network**


### droupout layer - randomly drop some % of neuron links
**bold text**


```python
 from keras.models import Sequential
 from keras.layers import Input,Dense,Dropout
```

```python
#Intialize the model
model = Sequential()

#Input layer
model.add(Input(shape=(X_train.shape[1], )))

#Hiden layer 1
model.add(Dense(64, activation='relu' ))

#add dropout layer 1
model.add(Dropout(0.4))

#Hiden layer 2
model.add(Dense(32, activation='relu' ))

#add dropout layer 2
model.add(Dropout(0.3))

#output layer
model.add(Dense(1, activation ='sigmoid'))

```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
hist = model.fit(X_train, Y_train, epochs=60, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(hist.history['loss'],label = 'Training Loss')
plt.plot(hist.history['val_loss'],label = 'Validation Loss')
plt.xlabel('epochs')
plt.ylabel('Binary cross entropy Loss')
plt.title('Learning curve')
plt.legend()
plt.show
```

```python
model.evaluate(X_train,Y_train)
```

```python
model.evaluate(X_test,Y_test)
```
